```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/Aquifer_Petrignano.csv')
df
```

```python
ts = df[['Date', 'Depth_to_Groundwater_P24']]
ts
```

```python
ts.dropna()
```

```python
ts = ts.set_index('Date')
ts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(figsize = (10, 5))
sns.lineplot(data = ts)
plt.tight_layout()
plt.show()
```

```python
ts =ts.dropna()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(ts, period = 365)
decomposition
```

```python
trend = decomposition.trend

fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = ts)
sns.lineplot(data = trend)

plt.tight_layout()

plt.show()
```

```python
seasonal = decomposition.seasonal

fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = ts)
sns.lineplot(data = seasonal)

plt.tight_layout()

plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    print("Resultados de la prueba de Dickey-Fuller:")
    dftest = adfuller(timeseries, autolag = "AIC")
    dfoutput = pd.Series(dftest[0:4], index = ["Test Statistic", "p-value", "#Lags Used", "Number of Observations Used"])
    for key,value in dftest[4].items():
        dfoutput["Critical Value (%s)"%key] = value
    return dfoutput

test_stationarity(ts)
```

Observación:
- El p-value da como resultado 0,071 lo cual se descarta que la serie sea estacionaria (<=0,05)

```python
residual = decomposition.resid

fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = ts)
sns.lineplot(data = residual)

plt.tight_layout()

plt.show()
```

Observación:
- La serie original muestra una mayor amplitud de variación debido a la estacionalidad

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(ts)

plt.tight_layout()

plt.show()
```

Observación:
- La autocorrelación se altamente significativa.
- Hay un cierto declive, quizás debido por evaporamiento o aumento del cuerpo de agua.

```python
ts_stationary = ts.diff().dropna()

test_stationarity(ts_stationary)
```
